### Fonksiyonlar ( Functions )

````python
def not_hesaplama(ad, puanlar):
    ortalama = 0
    for puan in puanlar:
        ortalama += puan
    ortalama /= len(puanlar)
    
    if ortalama > 90:
        harf = 'A'
    elif ortalama > 80:
        harf = 'B'
    elif ortalama > 60:
        harf = 'C'
    elif ortalama > 50:
        harf = 'D'
    else:
        harf = 'F'
    
    print(f'sim: {ad}\nOrtalama: {ortalama:.2f}\nHarf notu: {harf}\n', end="-"*20)

not_hesaplama('Ahmet DENZ', (60, 100, 75))
````

````python
def carp(a=10, b=10):
    sonuc = a * b
    return sonuc

sonuc = carp()
print(sonuc)

print(carp(b=32, a=2))
````

````python
def carp(*args):
    sonuc = 1
    for say in args:
        sonuc *= say
    return sonuc

sonuc = carp(1, 2, 20, 4)
print(sonuc)

print(carp(*range(1, 10)))
````

````python
def siralama_yap(*args, **kwargs):
    for no, key in enumerate(kwargs):
        value = kwargs[key]
        print(f'No: {no:<10}Key: {key:^20}Value: {value:>10}')
    
insanlar = {
    'Ahmet': 'DENZ', 
    'Yaar': 'KEMAL',
    'Ali': 'TEOMAN',
    'Guido Van': 'ROSSUM'
}
siralama_yap(**insanlar)
        
````

## Nesne Ynelimli Programlama GR ( Beginning OOP )

````python
class Araba:
    def sr(self):
        print('Araba gidiyor..')
    
    def dur(self):
        print('Araba duruyor..')
        

araba = Araba()
araba.sr()
araba.dur()
araba.sr()
````

````python
class Araba:
    """
    Python dersi iin tanmladmz araba snfmz.
    -------------------------------------------------
    Snf ierisinde arabann yapmasn istediimiz 
    ve yapabilecei fonksiyonlar barndryor.
    """
    def __init__(self, marka, model, yl, renk):
        """
        Bir nesne olutururken aadaki parametreleri
        vermeyi unutmayn!
        => Marka, Model, Yl ve Renk
        """
        self.marka = marka
        self.model = model
        self.yl = yl
        self.renk = renk
        self.durum = 0
        """
        0: Kapal
        1: alr halde duruyor
        2: Gidiyor
        """
        
    def altr(self):
        self.durum = 1 if not self.durum else 1
    
    def durdur(self):
        self.durum = 0 if self.durum else 1
        
    def sr(self):
        if self.durum:
            print('{} gidiyor'.format(self.model))
            self.durum = 2
        else:
            print('nce arabay altr')
    
    def dur(self):
        if self.durum:
            print('{} duruyor'.format(self.model))
            self.durum = 1
        else:
            print('nce arabay altr')
            
    def zellikleri_yazdr(self):
        print(f"Marka: {self.marka:<10}\
        Model: {self.model:<10}\
        Yl: {self.yl:<10}\
        Renk: {self.renk:<10}")
````

````python
help(Araba)

araba1 = Araba('Dodge', 'Challenger', '2016', 'Siyah')
araba1.sr()
araba1.altr()
araba1.sr()
araba1.zellikleri_yazdr()
araba1.dur()
````

````python
araba2 = Araba('Ford', 'Mustang', '1971', 'Sar-Siyah')
araba2.sr()
araba2.altr()
araba2.sr()
araba2.zellikleri_yazdr()
araba2.dur()
````
